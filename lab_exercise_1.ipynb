```python
# Perceptron for AND Gate

def step(x):
    if x >= 0:
        return 1
    else:
        return 0

def AND_perceptron(x1, x2):
    w1 = 1
    w2 = 1
    bias = -1.5

    summation = x1*w1 + x2*w2 + bias
    return step(summation)

inputs = [(0,0),(0,1),(1,0),(1,1)]

print("AND Gate Output")
for x1,x2 in inputs:
    print(x1,x2,"->",AND_perceptron(x1,x2))
```

```python
# Perceptron for NAND Gate

def step(x):
    if x >= 0:
        return 1
    else:
        return 0

def NAND_perceptron(x1, x2):
    w1 = -1
    w2 = -1
    bias = 1.5

    summation = x1*w1 + x2*w2 + bias
    return step(summation)

inputs = [(0,0),(0,1),(1,0),(1,1)]

print("NAND Gate Output")
for x1,x2 in inputs:
    print(x1,x2,"->",NAND_perceptron(x1,x2))
```
